# Start of tutorial after installing Pycaret (first code cell). After running the next cell, use +code to add a code cell to enter code from the Quickstart TutoriaL.

```python
# load sample dataset
from pycaret.datasets import get_data
data = get_data('diabetes')
```

```python
# Functional API

from pycaret.classification import *
s = setup(data, target = 'Class variable', session_id = 123)
```

```python
# OOP API

from pycaret.classification import ClassificationExperiment
s = ClassificationExperiment()
s.setup(data, target = 'Class variable', session_id = 123)
```

```python
# functional API
best = compare_models()

# OOP API
best = s.compare_models()
```

```python
print(best)
```

```python
# Analyze Model
# functional API
evaluate_model(best)

# OOP API
s.evaluate_model(best)
```

```python
# functional API
plot_model(best, plot = 'auc')

# OOP API
s.plot_model(best, plot = 'auc')
```

```python
# functional API
plot_model(best, plot = 'confusion_matrix')

# OOP API
s.plot_model(best, plot = 'confusion_matrix')
```

```python
# Predictions
# functional API
predict_model(best)

# OOP API
s.predict_model(best)
```

```python
# functional API
predictions = predict_model(best, data=data)
predictions.head()

# OOP API
predictions = s.predict_model(best, data=data)
predictions.head()
```

```python
# Save the model

# functional API
save_model(best, 'my_best_pipeline')

# OOP API
s.save_model(best, 'my_best_pipeline')
```

```python
# To load the model back in environment:

# functional API
loaded_model = load_model('my_best_pipeline')
print(loaded_model)

# OOP API
loaded_model = s.load_model('my_best_pipeline')
print(loaded_model)
```

# **Regression**

```python
# load sample dataset
from pycaret.datasets import get_data
data = get_data('insurance')
```

```python
from pycaret.regression import *
s = setup(data, target = 'charges', session_id = 123)
```

```python
from pycaret.regression import RegressionExperiment
s = RegressionExperiment()
s.setup(data, target = 'charges', session_id = 123)
```

```python
# functional API
best = compare_models()

# OOP API
best = s.compare_models()
```

```python
print(best)
```

```python
# functional API
evaluate_model(best)

# OOP API
s.evaluate_model(best)
```

```python
# functional API
plot_model(best, plot = 'residuals')

# OOP API
s.plot_model(best, plot = 'residuals')
```

```python
# functional API
plot_model(best, plot = 'feature')

# OOP API
s.plot_model(best, plot = 'feature')
```

```python
# functional API
predict_model(best)

# OOP API
s.predict_model(best)
```

```python
# functional API
predictions = predict_model(best, data=data)
predictions.head()

# OOP API
predictions = s.predict_model(best, data=data)
predictions.head()
```

```python
# functional API
save_model(best, 'my_best_pipeline')

# OOP API
s.save_model(best, 'my_best_pipeline')
```

```python
# functional API
loaded_model = load_model('my_best_pipeline')
print(loaded_model)

# OOP API
loaded_model = s.load_model('my_best_pipeline')
print(loaded_model)
```
